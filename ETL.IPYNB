## BẢNG ORDERS

```python

df_olist_orders_dataset.head(10)
```

```python
df_olist_orders_dataset.info()
```

```python
# Chuyển đổi định dạng dữ liệu cho cột ngày tháng
df_olist_orders_dataset['order_purchase_timestamp'] = pd.to_datetime(df_olist_orders_dataset['order_purchase_timestamp'])
df_olist_orders_dataset['order_approved_at'] = pd.to_datetime(df_olist_orders_dataset['order_approved_at'])
df_olist_orders_dataset['order_delivered_carrier_date'] = pd.to_datetime(df_olist_orders_dataset['order_delivered_carrier_date'])
df_olist_orders_dataset['order_delivered_customer_date'] = pd.to_datetime(df_olist_orders_dataset['order_delivered_customer_date'])
df_olist_orders_dataset['order_estimated_delivery_date'] = pd.to_datetime(df_olist_orders_dataset['order_estimated_delivery_date'])

```

```python

# Tính thời gian xử lý đơn hàng
df_olist_orders_dataset['processing_time'] = df_olist_orders_dataset['order_approved_at'] - df_olist_orders_dataset['order_purchase_timestamp']
# Tính thời gian giao hàng
df_olist_orders_dataset['delivery_time'] = df_olist_orders_dataset['order_delivered_customer_date'] - df_olist_orders_dataset['order_delivered_carrier_date']
```

```python
# Phân loại thời gian xử lý đơn hàng
bins = [pd.Timedelta(days=-1), pd.Timedelta(days=0), pd.Timedelta(days=1), pd.Timedelta(days=5), pd.Timedelta(days=10), pd.Timedelta(days=20), pd.Timedelta(days=100)]  # Sử dụng một giá trị đủ lớn để đại diện cho vô hạn
labels = ['error', 'less than 1 day', '1 to 5 days', '5 to 10 days ', '10 to 20 days', ' 20 days and above ']

# Sử dụng pd.cut để phân loại
df_olist_orders_dataset['processing_time_category'] = pd.cut(df_olist_orders_dataset['processing_time'], bins=bins, labels=labels, include_lowest=True)
df_olist_orders_dataset['processing_time_category'] = df_olist_orders_dataset['processing_time_category'].astype('object')
```

```python
# Tính toán giá trị cho cột 'Early Delivery'
df_olist_orders_dataset['Early Delivery'] =  df_olist_orders_dataset['order_estimated_delivery_date'] - df_olist_orders_dataset['order_delivered_customer_date']

# Tính toán giá trị cho cột 'Late Delivery'
df_olist_orders_dataset['Late Delivery'] = df_olist_orders_dataset['order_delivered_customer_date'] - df_olist_orders_dataset['order_estimated_delivery_date']

# Thay thế các giá trị không phù hợp với yêu cầu bằng 'null'
df_olist_orders_dataset['Early Delivery'] = df_olist_orders_dataset['Early Delivery'].apply(lambda x: x if x.days > 0 else pd.NaT)
df_olist_orders_dataset['Late Delivery'] = df_olist_orders_dataset['Late Delivery'].apply(lambda x: x if x.days > 0 else pd.NaT)
```

```python
# Hàm xác định trạng thái giao hàng
def determine_delivery_status(row):
    if pd.isnull(row['Early Delivery']) and pd.isnull(row['Late Delivery']):
        return np.nan
    elif pd.isnull(row['Early Delivery']) and row['Late Delivery'] == pd.Timedelta(0):
        return 'on time'
    elif pd.isnull(row['Late Delivery']) and row['Early Delivery'] == pd.Timedelta(0):
        return 'on time'
    elif pd.notnull(row['Early Delivery']):
        return 'early'
    elif pd.notnull(row['Late Delivery']):
        return 'late'
    else:
        return np.nan

# Áp dụng hàm để tạo cột mới
df_olist_orders_dataset ['delivery_status'] = df_olist_orders_dataset .apply(determine_delivery_status, axis=1)
```
